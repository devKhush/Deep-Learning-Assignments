# DataLoaders

```python
class AudioDataset(SPEECHCOMMANDS):
    def __init__(self, subset: str = None):
        super().__init__("./", download=True)

        def load_list(filename):
            filepath = os.path.join(self._path, filename)
            with open(filepath) as fileobj:
                return [os.path.normpath(os.path.join(self._path, line.strip())) for line in fileobj]

        if subset == "validation":
            self._walker = load_list("validation_list.txt")
        elif subset == "testing":
            self._walker = load_list("testing_list.txt")
        elif subset == "training":
            excludes = load_list("validation_list.txt") + load_list("testing_list.txt")
            excludes = set(excludes)
            self._walker = [w for w in self._walker if w not in excludes]


audio_labels = ['backward', 'bed', 'bird', 'cat', 'dog', 'down', 'eight', 'five', 'follow', 'forward',
                'four', 'go', 'happy', 'house', 'learn', 'left', 'marvin', 'nine', 'no', 'off', 'on',
                'one', 'right', 'seven', 'sheila', 'six', 'stop', 'three', 'tree', 'two', 'up', 'visual', 'wow', 'yes', 'zero']

audio_class = {k:v for k, v in zip(audio_labels, range(len(audio_labels)))}

audio_train_set = AudioDataset("training")
audio_test_set = AudioDataset("testing")

waveform, sample_rate, label, speaker_id, utterance_number = audio_train_set[0]
print(waveform.shape, sample_rate, label, speaker_id, utterance_number)
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

batch_size = 4

image_trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)

image_testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

ImageTrainLoader = torch.utils.data.DataLoader(image_trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)
ImageTestLoader = torch.utils.data.DataLoader(image_testset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)

dataiter = iter(ImageTrainLoader)
images, labels = next(dataiter)

print(images.shape, labels)
```

# Architecture

```python
class ANetwork(nn.Module):
    def __init__(self,):
        super().__init__()
        self.L1 = nn.Conv1d(1,12,3)
        self.L2 = nn.Conv1d(12,24,3)
        self.L3 = nn.Conv1d(24,48,3)

        self.POOL = nn.AdaptiveAvgPool1d(512)

        self.Linear = nn.Sequential(nn.Linear(512, 1024),
                                    nn.Linear(1024, 2048))

        self.FL = nn.Linear(2048, 35)

    def forward(self, INPUT):
        O = self.L1(INPUT)
        O = self.L2(O)
        O = self.L3(O)
        O = self.POOL(O)

        O = O.mean(1)

        O = self.Linear(O)
        O = self.FL(O)
        return O

AudioNet = ANetwork()
```

```python
class INetwork(nn.Module):
    def __init__(self,):
        super().__init__()
        self.L1 = nn.Conv2d(3,12,3)
        self.L2 = nn.Conv2d(12,24,3)
        self.L3 = nn.Conv2d(24,48,3)

        self.POOL = nn.AdaptiveAvgPool2d(24)

        self.Linear = nn.Sequential(nn.Linear(576, 1024),
                                    nn.Linear(1024, 2048))

        self.FL = nn.Linear(2048, 10)

    def forward(self, INPUT):
        O = self.L3(self.L2(self.L1(INPUT)))

        O = self.POOL(O)

        O = O.flatten(2).mean(1)

        O = self.Linear(O)
        O = self.FL(O)
        return O

ImageNet = INetwork()
```

# Loss Function

```python
Loss = nn.MSELoss()
```

# Training Pipeline

```python
# Training Hyperparameters
BS = 64
EPOCHS = 4
LR = 0.001
MOMENTUM = 0.9
```

## Audio Data Training

```python
# collate function
def collate_fn(batch):
    data, target = [], []

    for i in batch:
        aud, _, label, _, _ = i
        data.append(aud[:,:1000])
        target.append(torch.tensor(audio_class[label]))

    return torch.stack(data), torch.stack(target).to(dtype=torch.float32)



AudioDataLoader = torch.utils.data.DataLoader(dataset=audio_train_set,
                                              batch_size=BS,
                                              shuffle=True,
                                              collate_fn=collate_fn)

Optim = torch.optim.SGD(AudioNet.parameters(), lr=LR, momentum=MOMENTUM)

for E in range(EPOCHS):
    EL, EA = 0,0

    print("Total Minibatches: ", len(AudioDataLoader))

    for idx, minibatch in enumerate(AudioDataLoader):

        Optim.zero_grad()
        d, t = minibatch
        logits = AudioNet(d)
        l = Loss(
            logits.softmax(1).argmax(1).to(torch.float32),
            t.requires_grad_(True)
            )
        l.backward()
        Optim.step()

        acc = logits.softmax(1).argmax(1).eq(t).sum().div(BS).mul(100)

        print(f"Epoch: {E}, Minibatch: {idx}, Loss: {l.item():.2f}, Accuracy: {acc.item():.2f}")

        EL += l.item()
        EA += acc.item()

        # We are running for 20 minibatches
        if idx == 20:
            break

    print("--"*20)
    print(f"Epoch: {E}, Loss: {EL/20}, Accuracy: {EA/20}")
    print("--"*20)

    torch.save(
        obj={
            "network": AudioNet.state_dict(),
            "epoch": E
        },
        f=f"AudioNet_Epoch_{E}.pth"
    )
```
